```python
def feature_extraction(file_path):
    # load the audio file
    x, sample_rate = librosa.load(file_path, res_type='kaiser_fast')
    # extract features from the audio
    mfcc = np.mean(librosa.feature.mfcc(y=x, sr=sample_rate, n_mfcc=50).T, axis=0)
    
    return mfcc
```

```python
features = {}
original_directory = '../audio_data/'
directory1 = '../dataset_creation/output_datasets/dash/'
directory2 = '../dataset_creation/output_datasets/dot/'

for audio in os.listdir(original_directory):
    audio_path = original_directory+audio
    print(audio_path)
    features[audio_path] = feature_extraction(audio_path)
for audio in os.listdir(directory1):
    audio_path = directory1+audio
    print(audio_path)
    features[audio_path] = feature_extraction(audio_path)
for audio in os.listdir(directory2):
    audio_path = directory2+audio
    print(audio_path)
    features[audio_path] = feature_extraction(audio_path)
```

```python
# from sklearn import svm   # imported at the start

audio_path = '../audio_data/dash.wav'  # as sound 1
feature_array_dash_org = np.array(features[audio_path])

feature_array_dash_datasets = []
for audio in os.listdir(directory1):
    audio_path = directory1+audio
    feature_array_dash_datasets.append(np.array(features[audio_path]))
    
    
audio_path = '../audio_data/dot.wav'  # as sound 2
feature_array_dot_org = np.array(features[audio_path])

feature_array_dot_datasets = []
for audio in os.listdir(directory2):
    audio_path = directory2+audio
    feature_array_dot_datasets.append(np.array(features[audio_path]))


X_train = np.vstack([feature_array_dash_org, feature_array_dash_datasets, feature_array_dot_org, feature_array_dot_datasets])

# there are 40 samples each + 1 original
y_sound1 = np.zeros(41, dtype=np.int32)  
y_sound2 = np.ones(41, dtype=np.int32)  

y_train = np.concatenate([y_sound1, y_sound2])

# Assuming you have your training data X_train and y_train from the previous example
# Initialize SVM classifier
classifier = svm.SVC(kernel='linear')

# Train the classifier (assuming X_train and y_train are available)
classifier.fit(X_train, y_train)

```
